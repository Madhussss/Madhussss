```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Movies = []

try:
    source = requests.get('https://www.imdb.com/chart/top/')
    source.raise_for_status()
    
    soup = BeautifulSoup(source.text,'html.parser')
    
    movies = soup.find('tbody', class_="lister-list").find_all('tr')
    
    for movie in movies:
        rank = movie.find('td', class_="titleColumn").get_text(strip=True)[0]
        name = movie.find('td', class_="titleColumn").a.text
        year = movie.find('td', class_="titleColumn").span.text
        rating = movie.find('td', class_="ratingColumn").strong.text
        
        print(rank, name, year, rating)
        Movies.append([rank, name, year, rating])
        
except Error as er:
    print(er)
```

```python
df = pd.DataFrame(Movies,columns= ['rank', 'name', 'year', 'rating'])
```
